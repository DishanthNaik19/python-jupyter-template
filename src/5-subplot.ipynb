### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (1st subplot)
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y, color='blue', label='y = sin(x)')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Bar chart (2nd subplot)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['blue', 'green', 'orange', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot (3rd subplot)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5, -5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Pie chart (4th subplot)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'gold'], startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()





```
